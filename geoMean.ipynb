```python
import numpy as np
import matplotlib.pyplot as plt

img = plt.imread('imagens/cameraman.tiff')

filter_size = 3
img_geo = np.zeros(img.shape)
for y in range(filter_size//2, img.shape[0]-filter_size//2):
    for x in range(filter_size//2, img.shape[1]-filter_size//2):
        # Pega todos os vizinhos
        neighborhood = img[y:y+filter_size, x:x+filter_size]
        # Mdia Geomtrica dos vizinhos
        img_geo[y,x] = np.prod(neighborhood) ** (1/filter_size**2)


plt.imshow(img, cmap='gray')
plt.show()

plt.imshow(img_geo, cmap='gray')
plt.show()
```
